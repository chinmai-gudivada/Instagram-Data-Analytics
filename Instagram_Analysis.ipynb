```python
import pandas as pd
import plotly.express as px
```

```python
data = pd.read_csv('/content/Instagram data.csv',
                   encoding='latin-1')

data
```

```python
data.info()
```

```python
px.scatter(data, x='Impressions', y='Likes',
           trendline='ols', size='Likes',
           title='Likes vs Impressions')
```

```python
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

import matplotlib.pyplot as plt
```

```python
text = ' '.join(i for i in data.Caption)

stopwords = set(STOPWORDS)

wordcloud = WordCloud(stopwords=stopwords, background_color='black').generate(text)

plt.figure(figsize=(12, 10))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```

```python
home = data['From Home'].sum()
hashtags = data['From Hashtags'].sum()
explore = data['From Explore'].sum()
other = data['From Other'].sum()

labels = ['From Home', 'From Hashtags', 'From Explore', 'From Other']
values = [home, hashtags, explore, other]
```

```python
px.bar(x=labels, y=values,
       title='Impression from various sources',
       color=labels)
```

```python
px.pie(values=values, names=labels,
       hole=0.5)
```
